```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/data_hw_2__upd_.xlsx')
df.head()
```

```python
df_fr = df[df['Country'] == 'France']
df_ger = df[df['Country'] == 'Germany']
```

```python
sns.distplot(df_fr['TotalPrice'])
sns.distplot(df_ger['TotalPrice'])
plt.legend(['France', 'Germany'])
plt.title('Плотность рапсределения объемов продаж')
```

Вывод: обе выборки имеют распределение стремящееся к нормальному

```python
np.var(df_fr['TotalPrice'])/np.var(df_ger['TotalPrice']) # 0.6947034587752559 < 4
```

```python
from scipy.stats import ttest_ind, mannwhitneyu

res_st = ttest_ind(df_fr['TotalPrice'], df_ger['TotalPrice'], equal_var=True, )
res_u = mannwhitneyu(df_fr['TotalPrice'], df_ger['TotalPrice'])
res_st, res_u
```

p-value в обоих тестах показал значение ниже уровня значимости в 0.05 -> отвергаем нулевую гипотезу
