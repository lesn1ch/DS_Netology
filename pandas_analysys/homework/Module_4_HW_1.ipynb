# Задание 1
Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')
```

```python
movies.head(20)
```

```python
ratings.head(20)
```

```python
movie_rating = movies[['movieId', 'title']].merge(ratings[['movieId', 'rating']], how='inner', on='movieId')
```

```python
# только оценки 5
five_star = movie_rating.loc[movie_rating['rating'] == 5]

# группируем, считаем пятерки, сортируем и выводим первый фильм
five_star.groupby(by=['movieId', 'title']).count().sort_values(by='rating', ascending=False).head(1)
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
power = pd.read_csv('/content/power.csv')
```

```python
# отфильрованный датафрейм
filtred_power = power[(power['quantity'] >= 0) &
                      (power['category'].isin([4, 12, 21])) &
                      (2005 <= power['year']) & (power['year'] <= 2010) &
                      (power['country'].isin(['Estonia', 'Latvia', 'Lithuania']))]
filtred_power['quantity'].sum() #суммарное потребление
```
